```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft
```

```python
# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows 환경
```

```python
# CSV 파일에서 데이터 로드
file_path_normal = 'ship_data/train/normal/STFMK-20201105-LW17-8125_20201125_174621_004.csv'
data_normal = pd.read_csv(file_path_normal, encoding='utf-8')  # 인코딩을 'utf-8'로 설정
```

```python
data_normal
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data_normal['x'], data_normal['y'], linestyle='-')
plt.title('데이터 시각화 (전체 행)', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# 데이터의 첫 100행 시각화
plt.figure(figsize=(10, 6))
plt.plot(data_normal['x'][:100], data_normal['y'][:100], linestyle='-')
plt.title('데이터 시각화 (첫 100행)', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# 300행부터 500행까지의 데이터 시각화
plt.figure(figsize=(10, 6))
plt.plot(data_normal['x'][300:800], data_normal['y'][300:800], linestyle='-')
plt.title('데이터 시각화 (300행에서 800행)', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# 1000행부터 1100행까지의 데이터 시각화
plt.figure(figsize=(10, 6))
plt.plot(data_normal['x'][1000:1100], data_normal['y'][1000:1100], linestyle='-')
plt.title('데이터 시각화 (1000행에서 1100행)', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# y 값에 대한 FFT 수행
y_fft = np.fft.fft(data_normal['y'])
n = len(data_normal['y'])
frequencies = np.fft.fftfreq(n, d=(data_normal['x'][1] - data_normal['x'][0]))  # 주파수 계산
```

```python
# FFT 결과 시각화
plt.figure(figsize=(12, 6))
plt.plot(frequencies[:n//2], np.abs(y_fft)[:n//2], linestyle='-')
plt.title('y 값의 FFT', fontsize=14)
plt.xlabel('주파수', fontsize=12)
plt.ylabel('크기', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# STFT 스펙트로그램 계산 및 시각화
f, t, Zxx = stft(data_normal['y'], fs=1/(data_normal['x'][1] - data_normal['x'][0]), nperseg=100)
plt.figure(figsize=(12, 6))
plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
plt.title('STFT 스펙트로그램', fontsize=14)
plt.xlabel('시간', fontsize=12)
plt.ylabel('주파수', fontsize=12)
plt.colorbar(label='크기')
plt.show()
```
